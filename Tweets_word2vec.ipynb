```
import io
data = pd.read_csv(io.BytesIO(uploaded['train.csv']))
```

```
data.head()
```

```
import nltk
nltk.download('punkt')
```

```
import string
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```
def text_process(mess):
    mess = [i for i in mess if i not in string.punctuation]
    mess = "".join(mess)
    return [i for i in mess.split(" ") if i.lower() not in stopwords.words('english')]
```

```
l=data["TweetText"].apply(text_process)
```

```
l
```

```
#transformer les listes en phrases
tweets_ph = []
for li in l:
  ph=""
  for mot in li:
    ph=ph + " " + mot
  li = ph
  tweets_ph.append(li)
```

```
#exemple
tweets_ph[13]
```

```
tweetsvec = [nltk.word_tokenize(title) for title in tweets_ph]
```

```
tweetsvec[0]
```

```
from gensim.models import Word2Vec
w2v = Word2Vec( tweetsvec, size = 300, min_count=1)
```

```
#on remarque que la similarité ne marche pas si bien (j'ai essayé d'autre dimension de l'espace mais toujours les memes resultats)
#ps : word2vec comprend le contexte des mots par rapport aux nombres de mots qui sont similaire dans le meme tweet
w2v.similar_by_word('Messi')
```

```
w2v.similar_by_word('BarackObama')
```

```
print (list(w2v.wv.vocab))
print (len(list(w2v.wv.vocab)))
```

```
#calculer le barycentre de chaque tweet dans l'espace de 500 dimensions pour pouvoir classifier ces tweets en fonction du contexte des mots
list_of_centers = []
for i in tweetsvec :
  a=[32]
  d=0
  for j in i:
    a=a+w2v.wv.get_vector(j) 
    d=d+1 
  e=a/d 
  list_of_centers.append(e)
```

```
from sklearn.model_selection import train_test_split

tweets_train, tweets_test, label_train, label_test = \
train_test_split(list_of_centers, data['Label'], test_size=0.2)

print (len(tweets_train), len(tweets_test), len(label_train), len(label_test))
```

```
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(tweets_train,label_train)
```

```
from sklearn.metrics import classification_report
pred_train1 =  neigh.predict(tweets_train)
print (classification_report(label_train,pred_train1))
```

```
pred_test1 =  neigh.predict(tweets_test)
print (classification_report(label_test,pred_test1))
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
tree_para = {'criterion':['gini','entropy'],'max_depth':[1,2,4,8,12,14,20,30,40,50,60,70,100]}
clf = GridSearchCV(DecisionTreeClassifier(), tree_para, cv=5)
clf.fit(tweets_train,label_train)
```

```
clf.best_params_
```

```
pred_train2 =  clf.predict(tweets_train)
print (classification_report(label_train,pred_train2))
```

```
pred_test2 =  clf.predict(tweets_test)
print (classification_report(label_test,pred_test2))
```
